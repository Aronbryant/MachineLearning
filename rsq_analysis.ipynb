```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("original_data.xls")
```

```python
lv_non=pd.value_counts(data['有无水流'])['无']
lv_move=pd.value_counts(data['有无水流'])['有']
fig =plt.figure(figsize=(6,5))
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
plt.bar(left=range(2),height=[lv_non,lv_move],width=0.4,alpha=0.8,color='skyblue')
plt.xticks([index for index in range(2)],['无','有'])
plt.xlabel('水流状态')
plt.ylabel('数量')
plt.title('不同状态水流记录')
plt.show()
plt.close()
```

```python
water=data['水流量']
fig =plt.figure(figsize=(5,8))
plt.boxplot(water,
           patch_artist=True,
           labels=['水流量'],
           boxprops={'facecolor':'lightblue'})
plt.title('水流分布箱线图')
plt.grid(axis='y')
plt.show()
```
