```python

import networkx as ne #导入建网络模型包，命名ne
import matplotlib.pyplot as mp #导入科学绘图包，命名mp
#erdos renyi graphy
rg=ne.erdos_renyi_graph(100,0.07)
ps=ne.shell_layout(rg)#布置框架
ne.draw(rg,ps,with_labels=False,node_size=40)
mp.show()
```

```python
import networkx as nx #导入
import matplotlib.pyplot as plt 
#来生成一个有10个节点，每次加入2条边的无标度网络
ba=nx.barabasi_albert_graph(1000, 1,seed=42)
ps=nx.shell_layout(ba)#布置框架
nx.draw(ba,ps,with_labels=False,node_size=20)
plt.show()
```

```python
WS = nx.random_graphs.watts_strogatz_graph(100, 2,0.3)
# circular layout
pos = nx.circular_layout(WS)
nx.draw(WS, pos, with_labels = False, node_size = 30)
plt.show()
```
